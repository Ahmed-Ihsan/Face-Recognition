```python
def load_dataset(dataset_path):
    """
    Load images, labels, and image paths from the specified dataset path.

    Args:
    dataset_path (str): Path to the dataset directory

    Returns:
    tuple: Numpy arrays of images, labels, and a list of image paths
    """
    images = []
    labels = []
    image_paths = []  # List to store the image paths

    for person_name in os.listdir(dataset_path):
        person_path = os.path.join(dataset_path, person_name)

        if os.path.isdir(person_path):
            for folder in tqdm(os.listdir(person_path), desc="Loading User Images", total=len(os.listdir(person_path))):
                folder_path = os.path.join(person_path, folder)
                try:
                    for image_name in os.listdir(folder_path):
                        image_path = os.path.join(folder_path, image_name)
                        image = cv2.imread(image_path)

                        # Check if the image was loaded successfully
                        if image is not None:
                            images.append(image)
                            labels.append(person_name)
                            image_paths.append(image_path)  # Append the image path
                except Exception as e:
                    print(f"Error loading images from {folder_path}: {e}")

    return np.array(images), np.array(labels), image_paths
```

```python
def display_sample_images(images, labels, num_samples=5):
    """
    Display a random sample of images with their corresponding labels.

    Args:
    images (numpy.ndarray): Array of images
    labels (numpy.ndarray): Array of corresponding labels
    num_samples (int): Number of images to display (default is 5)
    """
    plt.figure(figsize=(15, 10))  # Adjust the figure size if necessary

    # Select random sample indices
    sample_indices = random.sample(range(len(images)), num_samples)

    for i, idx in enumerate(sample_indices):
        image = images[idx]
        label = labels[idx]

        # Plot each image
        plt.subplot(1, num_samples, i + 1)
        plt.imshow(image)
        plt.title(f"Label: {label}")
        plt.axis('off')  # Hide axis

    plt.show()
```

```python
def preprocess_images(images, labels,path):
    """
    Preprocess images by detecting faces using DeepFace's extract_faces, cropping them by detected coordinates,
    resizing, and normalizing. Remove images with no faces detected.

    Args:
    images (numpy.ndarray): Array of input images
    labels (list): Corresponding labels for images

    Returns:
    numpy.ndarray: Array of preprocessed images
    list: Corresponding labels for preprocessed images
    """
    preprocessed_images = []
    preprocessed_labels = []

    for idx, image in tqdm(enumerate(images), desc="Preprocessing images"):
        try:
            # Detect and extract faces using DeepFace's extract_faces
            img1 = cv2.imread(path[idx])
            faces = DeepFace.represent(img1)  # Use the current image
            if len(faces) > 0:
                for face_dict in faces:

                    # Extract bounding box coordinates
                    x, y, width, height = face_dict["facial_area"]["x"], face_dict["facial_area"]["y"], face_dict["facial_area"]["w"], face_dict["facial_area"]["h"]
                    # Crop the face image using the coordinates
                    cropped_face = img1[y:y + height, x:x + width]  # Crop using (y, y+height, x, x+width)

                    # Normalize the cropped face
                    # cropped_face = cropped_face.astype('float32') / 255.0

                    # resize image
                    cropped_face = cv2.resize(cropped_face, IMAGE_SIZE)

                    # Append the normalized cropped face image and corresponding label to lists
                    try:
                      preprocessed_images.append(cropped_face)
                      preprocessed_labels.append(labels[idx])  # Keep the corresponding label

                      # Plot the cropped image
                      plt.imshow(cropped_face, vmin=0, vmax=1)  # Use grayscale colormap
                      plt.axis('off')  # Hide axis
                      plt.title(f"Label: {labels[idx]}")  # Display the corresponding label
                      plt.show()  # Show each image separately
                    except:
                      break
            else:
                print(f"No face detected in image {idx}. Skipping this image.")

        except ValueError as e:
            # If an error occurs during processing, skip this image
            print(f"Error processing image {idx}: {e}")

    return preprocessed_images, preprocessed_labels

def apply_histogram_equalization(images):
    """
    Apply histogram equalization to enhance image contrast.

    Args:
    image (numpy.ndarray): Input image

    Returns:
    numpy.ndarray: Contrast-enhanced image
    """

    image_processing =[]
    for image in images:
        ycrcb = cv2.cvtColor(image, cv2.COLOR_RGB2YCrCb)
        channels = cv2.split(ycrcb)
        cv2.equalizeHist(channels[0], channels[0])
        ycrcb = cv2.merge(channels)
        image_processing.append(cv2.cvtColor(ycrcb, cv2.COLOR_YCrCb2RGB))
    return image_processing

def apply_gaussian_blur(images, kernel_size=(5, 5)):
    """
    Apply Gaussian blur to reduce image noise.

    Args:
    image (numpy.ndarray): Input image
    kernel_size (tuple): Size of the Gaussian kernel

    Returns:
    numpy.ndarray: Blurred image

    """
    image_processing =[]
    for image in images:
        image_processing.append(cv2.GaussianBlur(image, kernel_size, 0))
    return image_processing

def apply_edge_detection(images):
    """
    Apply Canny edge detection to highlight image edges.

    Args:
    image (numpy.ndarray): Input image

    Returns:
    numpy.ndarray: Edge-detected image
    """
    image_processing =[]
    for image in images:
        gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
        edges = cv2.Canny(gray, 100, 200)
        image_processing.append(cv2.cvtColor(edges, cv2.COLOR_GRAY2RGB))
    return image_processing


def advanced_preprocess_data(images,label,path):
    """
    Apply advanced preprocessing techniques to the input images.

    Args:
    images (numpy.ndarray): Array of input images

    Returns:
    numpy.ndarray: Array of preprocessed images
    """
    preprocessed_images = []
    # for image in tqdm(images,desc="Advanced_preprocessing data"):
        # image = cv2.resize(image, IMAGE_SIZE)
    image , Lable =preprocess_images(images,label,path)
    # image = apply_histogram_equalization(image)
    # image = apply_gaussian_blur(image)
    # image = apply_edge_detection(image)
    # image = image.astype('float32') / 255.0
    preprocessed_images.append(image)
    return np.array(preprocessed_images), Lable
```

```python
# 2. Data Augmentation
def augment_data(images, labels):
    """
    Augment the dataset using various image transformations.

    Args:
    images (numpy.ndarray): Array of input images
    labels (numpy.ndarray): Array of corresponding labels

    Returns:
    tuple: Augmented images and labels
    """
    datagen = ImageDataGenerator(
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest'
    )

    augmented_images = []
    augmented_labels = []

    for image, label in tqdm(zip(images, labels),desc="Augment Data"):
        image = image.reshape((1,) + image.shape)
        for _ in range(5):  # Generate 5 augmented images for each original image
            augmented = datagen.flow(image, batch_size=1)[0]
            augmented_images.append(augmented[0])
            augmented_labels.append(label)

    return np.array(augmented_images), np.array(augmented_labels)
```

```python
# 3. Feature Extraction

def extract_hog_features(images):
    """
    Extract Histogram of Oriented Gradients (HOG) features from images.

    Args:
    images (numpy.ndarray): Array of input images

    Returns:
    numpy.ndarray: Array of HOG features
    """
    hog_features = []
    for image in tqdm(images,desc="Extract hog features",):
#         print(image)
        gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
        fd, _ = hog(gray, orientations=8, pixels_per_cell=(16, 16), cells_per_block=(1, 1), visualize=True)
        hog_features.append(fd)
    return np.array(hog_features)

def extract_lbp_features(images):
    """
    Extract Local Binary Pattern (LBP) features from images.

    Args:
    images (numpy.ndarray): Array of input images

    Returns:
    numpy.ndarray: Array of LBP features
    """
    lbp_features = []
    for image in tqdm(images,desc="Extract lbp features"):
        gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
        lbp = feature.local_binary_pattern(gray, P=8, R=1, method="uniform")
        hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, 10), range=(0, 9))
        lbp_features.append(hist)
    return np.array(lbp_features)


def detect_landmarks(image):
  """
  Detect facial landmarks using dlib's pre-trained model.

  Args:
  image (numpy.ndarray): Input image

  Returns:
  numpy.ndarray: Array of landmark coordinates, or None if no face is detected
  """
  # Initialize the face detector and landmark predictor
  detector = dlib.get_frontal_face_detector()
  predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")

  # Ensure the image is in 8-bit format
  if image.dtype != np.uint8:
      image = (image * 255).astype(np.uint8)

  # Convert the image to grayscale if it's not already
  if len(image.shape) == 3:
      gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
  else:
      gray = image

  # Detect faces in the grayscale image
  faces = detector(gray)

  if len(faces) > 0:
      # Get the landmarks for the first detected face
      landmarks = predictor(gray, faces[0])
      # Convert landmarks to numpy array
      landmarks_array = np.array([(p.x, p.y) for p in landmarks.parts()])
      return landmarks_array
  else:
      return None

def extract_landmark_features(images):
    """
    Extract facial landmark features from images.

    Args:
    images (numpy.ndarray): Array of input images

    Returns:
    numpy.ndarray: Array of landmark features
    """
    landmark_features = []
    for image in tqdm(images,desc="Extract landmark features"):
        landmarks = detect_landmarks(image)
        if landmarks is not None:
            landmark_features.append(landmarks.flatten())
        else:
            landmark_features.append(np.zeros(68*2))  # 68 landmarks, each with x and y coordinates
    return np.array(landmark_features)

def gabor_filter(image, frequency, theta):
    """
    Apply Gabor filter to an image.

    Args:
    image (numpy.ndarray): Input image
    frequency (float): Frequency of the sinusoidal component
    theta (float): Orientation of the Gabor filter

    Returns:
    numpy.ndarray: Filtered image
    """
    sigma = 1.0
    kernel = cv2.getGaborKernel((21, 21), sigma, theta, frequency, 0.5, 0, ktype=cv2.CV_32F)
    filtered = cv2.filter2D(image, cv2.CV_8UC3, kernel)
    return filtered

def extract_gabor_features(images):
    """
    Extract Gabor filter features from images.

    Args:
    images (numpy.ndarray): Array of input images

    Returns:
    numpy.ndarray: Array of Gabor features
    """
    gabor_features = []
    frequencies = [0.1, 0.5, 1.0]
    thetas = [0, np.pi/4, np.pi/2, 3*np.pi/4]

    for image in tqdm(images,desc="Extract gabor features"):
        features = []
        gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
        for freq in frequencies:
            for theta in thetas:
                filtered = gabor_filter(gray, freq, theta)
                features.append(np.mean(filtered))
                features.append(np.std(filtered))
        gabor_features.append(features)

    return np.array(gabor_features)


def extract_features_deepface(images, model_name='VGG-Face'):
  """
  Extract features from a list of images using DeepFace.

  Parameters:
      images (list): A list of images in NumPy array format.
      model_name (str): The name of the model to use. Default is 'VGG-Face'.
                        Available options: 'VGG-Face', 'Facenet', 'OpenFace', 'DeepFace', 'DeepID'

  Returns:
      features (list): A list of extracted feature vectors for each image.
  """
  try:
      features = []
      for image in tqdm(images, desc="Extracting DeepFace features"):
          # Convert image to RGB if it's not already
          if len(image.shape) == 2 or image.shape[2] == 1:  # Grayscale image
              image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
          elif image.shape[2] == 4:  # RGBA image
              image = cv2.cvtColor(image, cv2.COLOR_RGBA2RGB)

          image = cv2.resize(image, IMAGE_SIZE)
          # Ensure the image is a NumPy array
          if isinstance(image, np.ndarray):
              # Extract features using DeepFace
              feature_vector = DeepFace.represent(image, model_name=model_name, enforce_detection=False)
              # print(feature_vector[0]["embedding"])
              features.append(feature_vector[0]["embedding"])
          else:
              print(f"Image is not a NumPy array: {type(image)}")

      return features

  except Exception as e:
      print(f"Error occurred while extracting features: {e}")
      return None

def combine_features(hog_features, lbp_features,gabor_features, landmark_features=0, deepface_features=0, facenet_features=0):
    """
    Combine all extracted features into a single feature vector.

    Args:
    hog_features (numpy.ndarray): HOG features
    lbp_features (numpy.ndarray): LBP features
    landmark_features (numpy.ndarray): Facial landmark features
    gabor_features (numpy.ndarray): Gabor filter features
    deepface_features (numpy.ndarray): DeepFace features
    facenet_features (numpy.ndarray): FaceNet features

    Returns:
    numpy.ndarray: Combined feature vector
    """
    return np.concatenate([hog_features, lbp_features,gabor_features, landmark_features, deepface_features], axis=1)
    # return np.concatenate([hog_features, lbp_features, landmark_features, gabor_features, deepface_features, facenet_features], axis=1)
```

```python
def create_decision_tree():
    """
    Create a Decision Tree classifier.

    Returns:
    DecisionTreeClassifier: Initialized Decision Tree model
    """
    return DecisionTreeClassifier(random_state=RANDOM_STATE)

def create_svm(kernel='rbf', C=1.0, gamma='scale'):
    """
    Create a Support Vector Machine classifier.

    Args:
    kernel (str): Kernel type to be used in the algorithm
    C (float): Regularization parameter
    gamma (str or float): Kernel coefficient

    Returns:
    SVC: Initialized SVM model
    """
    return SVC(kernel=kernel, C=C, gamma=gamma, random_state=RANDOM_STATE, probability=True)

def create_knn(n_neighbors=5, weights='uniform', metric='minkowski'):
    """
    Create a K-Nearest Neighbors classifier.

    Args:
    n_neighbors (int): Number of neighbors to use
    weights (str): Weight function used in prediction
    metric (str): Distance metric to use for the tree

    Returns:
    KNeighborsClassifier: Initialized KNN model
    """
    return KNeighborsClassifier(n_neighbors=n_neighbors, weights=weights, metric=metric)

def create_mlp(hidden_layer_sizes=(100,), activation='relu', solver='adam', alpha=0.0001):
    """
    Create a Multi-Layer Perceptron classifier.

    Args:
    hidden_layer_sizes (tuple): The ith element represents the number of neurons in the ith hidden layer
    activation (str): Activation function for the hidden layer
    solver (str): The solver for weight optimization
    alpha (float): L2 penalty (regularization term) parameter

    Returns:
    MLPClassifier: Initialized MLP model
    """
    return MLPClassifier(hidden_layer_sizes=hidden_layer_sizes, activation=activation, solver=solver, alpha=alpha, random_state=RANDOM_STATE)

def create_random_forest(n_estimators=100, max_depth=None):
    """
    Create a Random Forest classifier.

    Args:
    n_estimators (int): Number of trees in the forest
    max_depth (int or None): Maximum depth of the trees

    Returns:
    RandomForestClassifier: Initialized Random Forest model
    """
    return RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth, random_state=RANDOM_STATE)

def create_gradient_boosting(n_estimators=100, learning_rate=0.1, max_depth=3):
    """
    Create a Gradient Boosting classifier.

    Args:
    n_estimators (int): Number of boosting stages to be run
    learning_rate (float): Learning rate shrinks the contribution of each tree
    max_depth (int): Maximum depth of the individual estimators

    Returns:
    GradientBoostingClassifier: Initialized Gradient Boosting model
    """
    return GradientBoostingClassifier(n_estimators=n_estimators, learning_rate=learning_rate, max_depth=max_depth, random_state=RANDOM_STATE)

def create_xgboost(n_estimators=100, learning_rate=0.1, max_depth=3):
    """
    Create an XGBoost classifier.

    Args:
    n_estimators (int): Number of boosting rounds
    learning_rate (float): Step size shrinkage used in update
    max_depth (int): Maximum tree depth

    Returns:
    XGBClassifier: Initialized XGBoost model
    """
    return XGBClassifier(n_estimators=n_estimators, learning_rate=learning_rate, max_depth=max_depth, random_state=RANDOM_STATE)

def create_lightgbm(n_estimators=100, learning_rate=0.1):
    """
    Create a LightGBM classifier.

    Args:
    n_estimators (int): Number of trees
    learning_rate (float): Learning rate

    Returns:
    LGBMClassifier: Initialized LightGBM model
    """
    return LGBMClassifier(n_estimators=n_estimators, learning_rate=learning_rate, random_state=RANDOM_STATE)

def create_catboost(iterations=100, learning_rate=0.1):
    """
    Create a CatBoost classifier.

    Args:
    iterations (int): Number of trees
    learning_rate (float): Learning rate

    Returns:
    CatBoostClassifier: Initialized CatBoost model
    """
    return CatBoostClassifier(iterations=iterations, learning_rate=learning_rate, random_state=RANDOM_STATE, silent=True)
```

```python
# 8. Hyperparameter Tuning
def perform_grid_search(X, y, model, param_grid, cv=5):
    """
    Perform grid search for hyperparameter tuning.

    Args:
    X (numpy.ndarray): Feature matrix
    y (numpy.ndarray): Target vector
    model: Model to tune
    param_grid (dict): Dictionary with parameters names as keys and lists of parameter settings to try as values
    cv (int): Number of folds in cross-validation

    Returns:
    GridSearchCV: Fitted GridSearchCV object
    """
    grid_search = GridSearchCV(model, param_grid, cv=cv, n_jobs=-1, verbose=1)
    grid_search.fit(X, y)
    return grid_search
```

```python
# 5. Model Training and Evaluation

def train_model(model, X_train, y_train, X_val, y_val, epochs=EPOCHS, batch_size=BATCH_SIZE):
    """
    Train a model on the given data.

    Args:
    model: The model to train
    X_train (numpy.ndarray): Training data
    y_train (numpy.ndarray): Training labels
    X_val (numpy.ndarray): Validation data
    y_val (numpy.ndarray): Validation labels
    epochs (int): Number of epochs to train
    batch_size (int): Batch size for training

    Returns:
    history: Training history
    """
    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=epochs,
        batch_size=batch_size,
        verbose=1
    )
    return history

def evaluate_model(model, X_test, y_test):
    """
    Evaluate a model on the test data.

    Args:
    model: The model to evaluate
    X_test (numpy.ndarray): Test data
    y_test (numpy.ndarray): Test labels

    Returns:
    dict: Dictionary containing evaluation metrics
    """
    print("-------------------------------------->",np.array(X_test).shape)

    # Predict using the model
    y_pred = model.predict(X_test)

    # Handle predictions for different types of models
    if isinstance(model, (Sequential, Model)):  # Neural networks (Keras, TensorFlow)
        y_pred_classes = np.argmax(y_pred, axis=1)  # Convert probabilities to class indices
    else:  # Classical models (SVM, Decision Tree, etc.)
        y_pred_classes = y_pred  # These models directly return class labels

    # Convert y_test to class indices if it's one-hot encoded
    y_test_classes = np.argmax(y_test, axis=1) if len(y_test.shape) > 1 else y_test

    # Initialize an empty dictionary to store the results
    results = {}

    # Define the scoring metrics and their corresponding functions
    scores = [
        {'name': 'accuracy_score', 'function': accuracy_score},
        {'name': 'f1_score', 'function': lambda y_true, y_pred: f1_score(y_true, y_pred, average='weighted')},
        {'name': 'roc_auc_score', 'function': lambda y_true, y_pred: roc_auc_score(y_true, y_pred, average='weighted')},
        {'name': 'confusion_matrix', 'function': confusion_matrix},
        {'name': 'classification_report', 'function': classification_report}
    ]

    # Loop through each score
    for score in scores:
        try:
            # Special handling for ROC AUC in multiclass classification
            if score['name'] == 'roc_auc_score':
                if len(np.unique(y_test_classes)) > 2:  # Multiclass case
                    result = roc_auc_score(y_test_classes, y_pred, multi_class='ovr', average='weighted')
                else:  # Binary case
                    result = roc_auc_score(y_test_classes, y_pred)
            else:
                # Calculate other scores
                result = score['function'](y_test_classes, y_pred_classes)

            # Store the result in the dictionary
            results[score['name']] = result
            print(f"Successfully calculated {score['name']}")

        except Exception as e:
            print(f"Error calculating {score['name']}: {str(e)}")
            results[score['name']] = f"Error: {str(e)}"

    # Print the results
    for name, result in results.items():
        print(f"{name}: {result}")

    return results
```

```python
# 6. Visualization Functions

def plot_training_history(history):
    """
    Plot the training history of a model.

    Args:
    history: Training history returned by model.fit()
    """
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()

def plot_confusion_matrix(cm, classes):
    """
    Plot the confusion matrix.

    Args:
    cm (numpy.ndarray): Confusion matrix
    classes (list): List of class names
    """
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

def visualize_model_predictions(model, X_test, y_test, class_names, num_samples=5):
    """
    Visualize model predictions on random test samples.

    Args:
    model: Trained model
    X_test (numpy.ndarray): Test data
    y_test (numpy.ndarray): Test labels
    class_names (list): List of class names
    num_samples (int): Number of samples to visualize
    """
    indices = np.random.choice(len(X_test), num_samples, replace=False)

    for idx in indices:
        image = X_test[idx]
        true_label = np.argmax(y_test[idx])
        pred = model.predict(np.expand_dims(image, axis=0))
        pred_label = np.argmax(pred)

        plt.figure(figsize=(6, 6))
        plt.imshow(image)
        plt.title(f"True: {class_names[true_label]}\nPredicted: {class_names[pred_label]}")
        plt.axis('off')
        plt.show()

def plot_2d_scatter(features, labels, title="2D Feature Space", xlabel="Component 1", ylabel="Component 2"):
    """
    Plot a 2D scatter plot of the transformed features.

    Args:
    features (numpy.ndarray): 2D features to plot
    labels (numpy.ndarray): Corresponding labels for coloring
    title (str): Title of the plot
    xlabel (str): X-axis label
    ylabel (str): Y-axis label
    """
    plt.figure(figsize=(10, 7))
    sns.scatterplot(x=features[:, 0], y=features[:, 1], hue=labels, palette='viridis', s=50, edgecolor='k')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend(title="Classes")
    plt.show()
```

```python
def face_recognition_pipeline(dataset_path, test_size=0.2, random_state=RANDOM_STATE):
    """
    Execute the complete face recognition pipeline.
    Args:
    dataset_path (str): Path to the dataset
    test_size (float): Proportion of the dataset to include in the test split
    random_state (int): Random state for reproducibility
    Returns:
    dict: Dictionary containing trained models and their performance metrics
    """
    # Load and preprocess data
    X, y, path = load_dataset(dataset_path)
    X, y = X[:150], y[:150]  # Ensure there is enough data for training
    print(X.shape, y.shape)

    # Label encode to y
    label_encoder = LabelEncoder()
    y_encoded = label_encoder.fit_transform(y)
    print(set(y_encoded))

    # Advanced preprocessing
    X_processed, y_processed = advanced_preprocess_data(X, y_encoded, path)

    # Perform data augmentation
    X_augmented, y_augmented = augment_data(X_processed[0], y_processed)  # Ensure proper data shape

    # Split the data
    X_train, X_test, y_train, y_test = train_test_split(X_augmented, y_augmented, test_size=test_size, random_state=random_state, stratify=y_augmented)

    # Extract features
    X_train_hog = extract_hog_features(X_train)
    X_test_hog = extract_hog_features(X_test)

    X_train_lbp = extract_lbp_features(X_train)
    X_test_lbp = extract_lbp_features(X_test)

    X_train_landmarks = extract_landmark_features(X_train)
    X_test_landmarks = extract_landmark_features(X_test)

    X_train_gabor = extract_gabor_features(X_train)
    X_test_gabor = extract_gabor_features(X_test)

    X_train_deepface = extract_features_deepface(X_train)
    X_test_deepface = extract_features_deepface(X_test)

    # Combine features
    X_train_combined = combine_features(X_train_hog, X_train_lbp, X_train_landmarks, X_train_gabor, X_train_deepface)
    X_test_combined = combine_features(X_test_hog, X_test_lbp, X_test_landmarks, X_test_gabor, X_test_deepface)

    # Step 2: Apply PCA for dimensionality reduction
    pca = PCA(n_components=150)  # For model training
    X_train_pca = pca.fit_transform(X_train_combined)
    X_test_pca = pca.transform(X_test_combined)

    # Step 3: Plot PCA results
    plot_2d_scatter(X_train_pca, y_train, title="PCA of Training Data", xlabel="Principal Component 1", ylabel="Principal Component 2")

    # Step 4: Apply t-SNE for visualization only
    tsne = TSNE(n_components=2, random_state=42)  # Reduce to 2D for visualization
    X_train_tsne = tsne.fit_transform(X_train_combined)

    # Step 5: Plot t-SNE results
    plot_2d_scatter(X_train_tsne, y_train, title="t-SNE of Training Data", xlabel="t-SNE Component 1", ylabel="t-SNE Component 2")

    # Encode labels
    y_train_encoded = label_encoder.fit_transform(y_train)
    y_test_encoded = label_encoder.transform(y_test)  # Transform using the same encoder
    num_classes = len(label_encoder.classes_)

    # Convert to categorical
    y_train_cat = to_categorical(y_train_encoded)
    y_test_cat = to_categorical(y_test_encoded)

    # Create and train models
    # Model creation dictionary
    models = {
        'Decision Tree': create_decision_tree(),
        'SVM': create_svm(),
        'KNN': create_knn(),
        'MLP': create_mlp(),
        'Random Forest': create_random_forest(),
        'Gradient Boosting': create_gradient_boosting(),
        'XGBoost': create_xgboost(),
        'LightGBM': create_lightgbm(),
        # 'CatBoost': create_catboost()
    }

    results = {}

    for name, model in models.items():
        print(f"\nTraining and evaluating {name}...")

        model.fit(X_train_combined, y_train_encoded)  # Train on PCA features

        metrics = evaluate_model(model, X_test_combined, y_test_encoded)  # Evaluate on PCA features
        results[name] = {
            'model': model,
            'metrics': metrics
        }

        print(f"Metrics for {name}:")
        print(metrics)

        plot_confusion_matrix(metrics['confusion_matrix'], label_encoder.classes_)

    # Visualize predictions for the best model
    best_model_name = max(results, key=lambda k: results[k]['metrics']['accuracy'])
    best_model = results[best_model_name]['model']
    visualize_model_predictions(best_model, X_test_combined, y_test_encoded, label_encoder.classes_)

    return results
```

```python
# 9. Model Deployment Functions

def save_model(model, filename):
    """
    Save a trained model to disk.

    Args:
    model: Trained model to save
    filename (str): Name of the file to save the model
    """
    model.save(filename)
    print(f"Model saved as {filename}")

def load_model(filename):
    """
    Load a trained model from disk.

    Args:
    filename (str): Name of the file to load the model from

    Returns:
    Model: Loaded model
    """
    return tf.keras.models.load_model(filename)

def predict_single_image(model, image_path, class_names):
    """
    Make a prediction on a single image.

    Args:
    model: Trained model
    image_path (str): Path to the image file
    class_names (list): List of class names

    Returns:
    tuple: Predicted class name and confidence score
    """
    image = cv2.imread(image_path)
    # image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image, IMAGE_SIZE)
    # image = image.astype('float32') / 255.0

    # Extract features (implement this based on your feature extraction pipeline)
    # X_test_hog = extract_hog_features(image)
    # X_test_lbp = extract_lbp_features(image)
    # X_test_gabor = extract_gabor_features(image)
    features = extract_features_deepface(image)
    print(np.expand_dims(features, axis=0))
    # Apply PCA
    pca = PCA(n_components=5)
    principal_components = pca.fit_transform(features)
    print("/n===================pca=====================/n",principal_components)
    # features = combine_features(X_test_hog, X_test_lbp, X_test_gabor,deepface_features = features)
    # features = extract_features_facenet(image)

    prediction = model.predict(np.expand_dims(principal_components, axis=0))
    predicted_class = np.argmax(prediction)
    confidence = np.max(prediction)

    return class_names[predicted_class], confidence
```
